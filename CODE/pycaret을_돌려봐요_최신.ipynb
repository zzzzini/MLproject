```python
def calculate_duration(df, date_column, reference_date_str='2024-02'):
    # Convert the reference date string to a datetime object
    reference_date = pd.to_datetime(reference_date_str + '-01', format='%Y-%m-%d')

    # Convert the date column to datetime objects
    df[date_column] = pd.to_datetime(df[date_column] + '-01', format='%Y-%m-%d')

    # Calculate the difference in months
    df['duration'] = ((reference_date.year - df[date_column].dt.year) * 12 +
                               reference_date.month - df[date_column].dt.month)
```

```python
calculate_duration(df, date_column='Date')
```

```python
import colorsys

# RGB를 HSV로 변환하는 함수
def rgb_to_hsv(row):
    r, g, b = row['R'], row['G'], row['B']
    h, s, v = colorsys.rgb_to_hsv(r / 255.0, g / 255.0, b / 255.0)
    return pd.Series({'H': h, 'S': s, 'V': v})

# DataFrame에 변환 함수를 적용
df[['H', 'S', 'V']] = df.apply(rgb_to_hsv, axis=1)

#  원래의 RGB 열을 삭제
df = df.drop(['R', 'G', 'B'], axis=1)

df.head(5)
```
