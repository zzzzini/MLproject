```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("sinmuhyung.jpeg", cv2.IMREAD_GRAYSCALE)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
hist = cv2.calcHist([image], [0], None, [256], [0, 256])

plt.subplot(2, 2, 2)

plt.imshow(image)
plt.axis('off')

plt.subplot(2, 2, 1)

plt.plot(hist)
plt.title('Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.show()

min, max, dummy1, dummy2 = cv2.minMaxLoc(hist)
contrast = (max - min) / (max + min)

print(f"Min Intensity: {np.min(image)}")
print(f"Max Intensity: {np.max(image)}")
print(f'최빈값: {np.argmax(hist)}')
print(f'평균 명암: {np.mean(image)}')
print(f'명암비: {contrast}')
```
