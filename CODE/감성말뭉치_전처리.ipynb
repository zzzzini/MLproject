```python
data = {'Sentence': df['사람문장1'],
        'emotion':  df['감정_대분류']}

df1 = pd.DataFrame(data)

df1
```

```python
data1 = {'Sentence': df['사람문장2'],
        'emotion':  df['감정_대분류']}

df2 = pd.DataFrame(data1)

df2
```

```python
print(df1['emotion'].unique())
print(df2['emotion'].unique())
```

```python
df3 =pd.concat([df1,df2], axis=0)
```
