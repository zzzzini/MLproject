# 필요한 요소 선언
--> 알아야 하는 정보 : Client ID, Client Secret

```python
import sys
import requests
import json

url = "https://naveropenapi.apigw.ntruss.com/sentiment-analysis/v1/analyze"

headers = {

    "X-NCP-APIGW-API-KEY-ID": "Client ID",
    "X-NCP-APIGW-API-KEY": "Client Secret",
    "Content-Type": "application/json"

}
```

```python
import pandas as pd
data = pd.read_csv("영상제목있는csv파일")
```

```python
li = data.loc[:,"Title"]
```

```python
gamsung = []
```

```python
for item in li:
  data = {"content": item}

  response = requests.post(url, data=json.dumps(data), headers=headers)
  rescode = response.status_code
  if(rescode == 200):
    gamsung.append(response.text)
    print(response.text)
  else:
      print("Error : " + response.text)
```

# 감정분석에 필요한 요소들만 정리

```python
s_data = []
```

```python
for item in gamsung:
  temp = json.loads(item)["document"]
  confidence = temp["confidence"]

  sentiment = temp["sentiment"]
  neg_confidence = confidence["negative"]
  pos_confidence = confidence["positive"]
  neu_confidence = confidence["neutral"]

  s_data.append([sentiment, neg_confidence, pos_confidence, neu_confidence])
```

```python
new_df = pd.DataFrame(s_data, columns=["sentiment", "neg_confidence", "pos_confidence", "neu_confidence"])
```
